```python
#assigning integer values to values in Reporting column
def trans_reporting(x):
    if x == 'Beginner':
        return 0
    if x == 'Intermediate':
        return 1
    if x == 'Advanced':
        return 2
```

```python
#assigning integer values to values in Assignee column
def trans_assignee(x):
    if x == 'Particular':
        return 0
    if x == 'Random':
        return 1
```

```python
#assigning integer values to values in Status column
def trans_status(x):
    if x == 'Unconfirmed':
        return 0
    if x == 'New':
        return 1
```

```python
#assigning integer values to values in Notification column
def trans_notification(x):
    if x == 'Normal':
        return 0
    if x == 'High':
        return 1
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

class RFCClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, n_estimators=100, random_state=None):
        self.model = RandomForestClassifier(n_estimators=n_estimators, random_state=random_state)
        self.n_estimators = n_estimators
        self.random_state = random_state

    def fit(self, X, y=None):
        self.model.fit(X, y)
        return self

    def predict(self, X):
        return self.model.predict(X)

# Initialize RFC classifier
rfc_classifier = RFCClassifier(n_estimators=100, random_state=None)
n_iterations = 10  # Number of iterations
k_folds = 10  # Number of folds

# Initialize a list to store cross-validation scores for RFC
rfc_cv_scores = []

for _ in range(n_iterations):
    # Perform cross-validation and calculate scores for RFC
    kf = KFold(n_splits=k_folds, shuffle=True, random_state=None)
    scores = cross_val_score(rfc_classifier, X, y, cv=kf)
    
    # Append the scores to the list for RFC
    rfc_cv_scores.append(scores)

# Convert the list of scores for RFC to a NumPy array for analysis
rfc_cv_scores = np.array(rfc_cv_scores)

# Print the cross-validation scores for each iteration and fold for RFC
for i, scores in enumerate(rfc_cv_scores):
    print(f"Iteration {i + 1}:")
    for j, score in enumerate(scores):
        print(f"Fold {j + 1}: {score:.4f}")
    print("")

# Calculate and print the mean and standard deviation of the scores for RFC
rfc_mean_scores = np.mean(rfc_cv_scores, axis=1)
rfc_std_scores = np.std(rfc_cv_scores, axis=1)
print("Mean Scores for RFC:", rfc_mean_scores)
print("Standard Deviations for RFC:", rfc_std_scores)
```
