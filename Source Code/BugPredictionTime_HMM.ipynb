```python
#assigning integer values to values in Reporting column
def trans_reporting(x):
    if x == 'Beginner':
        return 0
    if x == 'Intermediate':
        return 1
    if x == 'Advanced':
        return 2
```

```python
#assigning integer values to values in Assignee column
def trans_assignee(x):
    if x == 'Particular':
        return 0
    if x == 'Random':
        return 1
```

```python
#assigning integer values to values in Status column
def trans_status(x):
    if x == 'Unconfirmed':
        return 0
    if x == 'New':
        return 1
```

```python
#assigning integer values to values in Notification column
def trans_notification(x):
    if x == 'Normal':
        return 0
    if x == 'High':
        return 1
```

```python
from sklearn.model_selection import cross_val_score, KFold
from hmmlearn import hmm
from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np

class HMMClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, n_components=3, covariance_type="full"):
        self.model = hmm.GaussianHMM(n_components=n_components, covariance_type=covariance_type)
        self.n_components = n_components
        self.covariance_type = covariance_type

    def fit(self, X, y=None):
        lengths = [len(X)]
        self.model.fit(X, lengths=lengths)
        return self

    def predict(self, X):
        return self.model.predict(X)


# Initialize HMM classifier with number of components equal to unique values in y
hmm_classifier = HMMClassifier(n_components=len(np.unique(y)))

n_iterations = 10  # Number of iterations
k_folds = 10  # Number of folds

# Initialize a list to store cross-validation scores
cv_scores = []

for _ in range(n_iterations):
    # Create a KFold object for k-fold cross-validation
    kf = KFold(n_splits=k_folds, shuffle=True, random_state=None)
    
    # Perform cross-validation and calculate scores
    scores = cross_val_score(hmm_classifier, X, y, cv=kf)
    
    # Append the scores to the list
    cv_scores.append(scores)

# Convert the list of scores to a NumPy array for analysis
cv_scores = np.array(cv_scores)

# Print the cross-validation scores for each iteration and fold
for i, scores in enumerate(cv_scores):
    print(f"Iteration {i + 1}:")
    for j, score in enumerate(scores):
        print(f"Fold {j + 1}: {score:.4f}")
    print("")

# Calculate and print the mean and standard deviation of the scores
mean_scores = np.mean(cv_scores, axis=1)
std_scores = np.std(cv_scores, axis=1)
print("Mean Scores:", mean_scores)
print("Standard Deviations:", std_scores)
```
